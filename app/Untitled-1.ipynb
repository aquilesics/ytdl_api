```python
import youtube_dl
```

```python
class MyLogger(object):
    def debug(self, msg):
        pass

    def warning(self, msg):
        pass

    def error(self, msg):
        print(msg)


def my_hook(d):
    if d['status'] == 'finished':
        print('Done downloading, now converting ...')


ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': './/app//tmp//%(title)s.%(ext)s',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
    'logger': MyLogger(),
    'progress_hooks': [my_hook],
}
```

```python
def get_info(url):
    try:
        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            _info =  ydl.extract_info(url, download=False )        
        return _info
    except:
        return 'bad request!', 400
```

```python
filtro = [
        'categories','channel','chapters','display_id','duration'
        ,'extractor','webpage_url','webpage_url_basename','view_count','categories'
        ,'channel','chapters','display_id','duration','extractor','webpage_url'
        ,'webpage_url_basename','view_count','title','_type','thumbnail','entries'
    ]

def clear_info( _info ) -> dict:  
    clean_info = dict(filter (lambda x:x[0] in filtro, _info.items() )) 
    if "_type" in clean_info:
        clean_info["entries"] = list(map(lambda x:clear_info(x), clean_info["entries"]))
    
    return dict( clean_info )

```
